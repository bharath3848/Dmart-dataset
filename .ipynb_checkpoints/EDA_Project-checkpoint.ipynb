## Outlier detection 

### An outlier is an observation that lies an abnormal distance from other values in a random sample from a population.

```python
sns.boxplot(x = 'Price',data=df)            # sns.boxplot gives us the visualization of boxplot                
plt.show()
```

```python
sns.boxplot(x='DiscountedPrice',data=df)
plt.show()
```

### Here we can see there are few outliers in Price and DiscountedPrice columns 

```python
df['Price'].quantile([0.5,0.67,0.79,0.95,0.99])       # df.quantile gives the percentage of specific values
```

```python
df['Price'].max()                                    # df.max gives the maximum value of the column in a dataset
```

```python
df['DiscountedPrice'].quantile([0.5,0.67,0.79,0.95,0.99])        # df.quantile gives the percentage of specific values
```

```python
df['DiscountedPrice'].max()                           # df.max gives the maximum value of the column in a dataset
```

### As we can see there is huge difference between 99 percentile and maximum value of the particular column, So by this we can surely say there are outliers in it.

## Inter Quartile Range(IQR) Method

```python
Q1=df['Price'].quantile(0.25)                  # df.quantile(0.25) gives the 25th percentile of a column
Q3=df['Price'].quantile(0.75)                  # df.quantile(0.75) gives the 75th percentile of a column
Q1,Q3
```

```python
IQR = Q3-Q1                                   # IQR = Difference between 75th percentile and 25th percentile
IQR
```

```python
lower_limit = Q1 - 1.5*IQR                   # Lower limit and Upper limit to display the dataset without outliers.
upper_limit = Q3 + 1.5*IQR
lower_limit,upper_limit
```

```python
df_no = df[(df['Price'] > lower_limit) & (df['Price'] < upper_limit)]     # we display the datset without outliers
df_no
```

### So after applying IQR formula we removed outliers and we displayed the datset with no outliers in Price Column

```python
Q1=df['DiscountedPrice'].quantile(0.25)                     # df.quantile(0.25) gives the 25th percentile of a column
Q3=df['DiscountedPrice'].quantile(0.75)                     # df.quantile(0.75) gives the 75th percentile of a column
Q1,Q3
```

```python
IQR = Q3-Q1                                                 # IQR = Difference between 75th percentile and 25th percentile
IQR
```

```python
lower_limit1 = Q1 - 1.5*IQR                               # Lower limit and Upper limit to display the dataset without outliers.
upper_limit1 = Q3 + 1.5*IQR
lower_limit1,upper_limit1
```

```python
df_no_1 = df[(df['DiscountedPrice'] > lower_limit) & (df['DiscountedPrice'] < upper_limit)]
df_no_1
```

### So after applying IQR formula we removed outliers and we displayed the datset with no outliers in DiscountedPrice Column

```python
sns.boxplot(x='Price',data=df_no)                             # sns.boxplot gives us the visualization of boxplot 
plt.show()
```

```python
sns.boxplot(x='DiscountedPrice',data=df_no_1)
plt.show()
```

### Now we can clearly see by vissualization there are no outliers present .

# Univariate Analysis

## Categorical Univariate Analysis

## Countplot

### Countplot is basically a count of frequency plot in form of a bar graph. It plots the count of each category in a separate bar. When we use the pandasâ€™ value counts function on any column, It is the same visual form of the value counts function.

### Value counts gives the number of values in the choosen column,plot bar gives the bar graph of that and normalize function gives the percentage value,figsize adjusts the size of the plot

```python
df['Category'].value_counts(normalize=True).plot.barh(figsize=[10,7])
plt.show()
```

### By visualizing the countplot we have more products of Personal Care Category and less products of Smart Watches Category in Category Column

```python
df['SubCategory'].value_counts(normalize=True).plot.barh(figsize=[10,15])
plt.show()
```

### By visualizing the countplot we have more products of Snacks&Farsans and less products of Smart Watches  in SubCategory Column

```python
df['Brand'].value_counts(normalize=True).plot.barh(figsize=[30,200])  # value counts gives the number of values in the choosen column
plt.show()
```

### We filled the missing values by others in the Brand column if we visualize the countplot we have more products of unknown brands(i.e.,others) after that we have more products of Premia brand and less products of Candyman brand

```python
df['Name'].value_counts().head(30).plot.barh(figsize=[10,7])   # plot.bar shows the bar graph of the selected column 
plt.show()
```

### As Name column have many values I have taken some values by using head and visualized the bargraph in those values we have more number of products of a product named "Elle 18  Nail pops Nail colour "

## Numerical Univariate Analysis

## Histplot 

### A histogram is a value distribution plot of numerical columns.we can have a look where more values lie like in positive, negative, or at the center(mean).

### sns. histplot gives the distribution plot , kde = True gives us a KDE(Kernel Density Estimation) over histogram which explains PDF(Probability Density Function) which means what is the probability of each value occurring in this column. 

```python
sns.histplot(df_no_1['Price'],kde=True,linewidth=0) # sns.histplot gives the distribution plot,kde = True gives us a KDE(Kernel Density Estimation) over histogram
plt.show()
```

### By observing the above histogram we can say that the Price of most of the products are between 0 to 500 rupees and there are very less products of higher price 

```python
sns.histplot(df_no_1['DiscountedPrice'],kde=True,linewidth=0) # sns.histplot gives the distribution plot,kde = True gives us a KDE(Kernel Density Estimation) over histogram
plt.show()
```

### By observing the above histogram we can say that the DiscountedPrice of most of the products are between 0 to 150 rupees and there are very less products of higher Discountedprice

## Distplot

### Distplot explains the PDF function using kernel density estimation

```python
sns.distplot(df_no_1[df_no_1['Brand'] == 'Navneet']['Price'],kde=True,hist=False,color="blue",) 
sns.distplot(df_no_1[df_no_1['Brand'] == 'Premia']['Price'],kde=True,hist=False,color="orange")
plt.show()
```

### As we can see the graph is really very interesting. the blue one shows the price of brand Navneet and the orange plot shows the price of brand premia. If we observe it we can see that price of Navneet is higher than premia.
